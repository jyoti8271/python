```python
## nested conditional statements
#you can place one or more if,elif,or else statements inside anither if,elif,or else statement to created nested conditional statements
#number,even,odd,negative

num=int(input("enter the number"))
 
if num > 0:
    print("the number is positive")
    if num%2 == 0:
          print("number is even")
    else:
        print("number is odd")    
else:
    print("negative or equal to zero")        
```

```python
## practical examples:

##you need to deterimine if a year is a leap year using nested condtion statement
year=int(input("enter the year"))
if year%4==0:
    if year%100==0:
       if year%400==0:
           print("leap year",year)
       else:
           print("not leap year",year)
    else:
        print("leap year")
else:
    print("not a leap year",year)                              
```

```python
##Assignmemt 
# simple calculator 
num1=float(input("enter the first number"))
num2=float(input("enter the second number"))
operation= input("enter the operation(+,-,*,/): ")

#perform the requested operations
if operation =="+":
    result =num1+ num2
elif operation=="-":
    result=num1-num2   
elif operation=="*":
    result =num1* num2
elif operation=="/":
    if num2 != 0:    
      result =num1/ num2 
    else:
        result ="error"
else:
    print("invalid operatio")

print("result is",result)                     
```
