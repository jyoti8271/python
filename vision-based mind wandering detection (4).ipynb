# Time based sampling

```python
def load_video(video_path, num_frames=24):
    frames = []

    try:
        cap = cv2.VideoCapture(video_path)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fps = cap.get(cv2.CAP_PROP_FPS)

        if total_frames == 0 or fps == 0:
            return None

        # sample one frame every 0.5 seconds
        step = int(fps * 0.5)

        for i in range(0, total_frames, step):
            cap.set(cv2.CAP_PROP_POS_FRAMES, i)
            ret, frame = cap.read()
            if not ret:
                continue

            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame = cv2.resize(frame, (224, 224))
            frames.append(frame)

            if len(frames) == num_frames:
                break

        cap.release()
        return np.array(frames)

    except Exception as e:
        print(f"Error reading {video_path}: {e}")
        return None
```

# Step-4 : Face Cropped Frames(Boundary Box Method)

```python
os.makedirs("cropped_faces", exist_ok=True)
cropped_faces = []

def extract_faces(frames, mtcnn, min_face_size=40):
    face_count = 0

    for idx, frame in enumerate(frames):
        try:
            frame = frame.astype(np.uint8)

            boxes, probs = mtcnn.detect(frame)

            if boxes is None:
                print(f"No face in frame {idx}")
                continue

            x1, y1, x2, y2 = boxes[0].astype(int)

            h, w, _ = frame.shape
            x1, y1 = max(0, x1), max(0, y1)
            x2, y2 = min(w, x2), min(h, y2)

            if (x2 - x1) < min_face_size or (y2 - y1) < min_face_size:
                print(f"Face too small in frame {idx}")
                continue

            face_crop = frame[y1:y2, x1:x2]

            if face_crop.size == 0:
                print(f"Empty crop in frame {idx}")
                continue

            face_crop = cv2.resize(face_crop, (160, 160))

            #save_path = f"cropped_faces/_{face_count}.jpg" --->Not Recommended, it'll create issues during training, crash 
            #cv2.imwrite(save_path, cv2.cvtColor(face_crop, cv2.COLOR_RGB2BGR))

            cropped_faces.append(face_crop)
            face_count += 1

        except Exception as e:
            print(f"Skipped Frame {idx}: {e}")

    if len(cropped_faces) == 0:
        return None

    return np.array(cropped_faces)
```

# Step-6 : CNN + Attention-Based MIL

### dataset --> mil dataset
#### (frame, label) --> (frames_of_clip, clip_label)

```python
import re
def extract_number(filename):
    return int(re.search(r'\d+',filename).group())
    
mil_dataset = []

for clip_id in os.listdir(save_root):
    clip_folder = os.path.join(save_root, clip_id)
    if not os.path.isdir(clip_folder):
        continue

    label = label_dict.get(clip_id)
    if label is None:
        continue

    #collecting all frames for this clip
    frame_paths = [
        os.path.join(clip_folder, f)
        for f in sorted(os.listdir(clip_folder), key=extract_number)
        if f.lower().endswith((".jpg", ".png", ".jpeg"))
    ]

    #skip empty clip
    if len(frame_paths) == 0:
        continue

    mil_dataset.append((frame_paths, label)) #--> takes tupple : (frames_of_clip,label)

print(f"Total Clips: {len(mil_dataset)}")
print(f"Sample bag: {mil_dataset[0][0][:6]} , Label: {mil_dataset[0][1]}")
```

# CNN Feature extractor

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

resnet = models.resnet18(pretrained=True)
resnet.fc = nn.Identity() # Remove classifier
resnet = resnet.to(device)
resnet.eval()
```

### Image Preprocessing

```python
#Load & preprocess all frames

transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(), # PIL format --> PyTorch tensor
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

### Convert Frames → Feature Bag (T, 512)

```python
def extract_clip_features(frame_paths):
    features = []

    with torch.no_grad():
        for path in frame_paths:
            img = Image.open(path).convert("RGB")
            img = transform(img).unsqueeze(0).to(device)  # (1,3,224,224)

            feat = resnet(img)  # (1,512)
            features.append(feat.squeeze(0))

    return torch.stack(features)
```

```python
# Load all frames of one clip -->Model treats this as ONE training sample
frame_paths, label = mil_dataset[0]
```

### One-Time CNN Feature Extraction

```python
os.makedirs("features", exist_ok=True)
```

```python
import numpy as np

print(len(mil_dataset))
for clip_id, (frame_paths, label) in enumerate(mil_dataset):

    save_path = f"features/{clip_id}.npy"

    # Skip if already extracted
    if os.path.exists(save_path):
        continue
        

    bag_features = extract_clip_features(frame_paths)  # (T,512)

    np.save(save_path, bag_features.cpu().numpy())
    
```

### Create Final MIL Dataset

```python
final_mil_dataset = []

for clip_id, (_, label) in enumerate(mil_dataset):
    feature_path = f"features/{clip_id}.npy"

    bag = torch.tensor(np.load(feature_path))
    final_mil_dataset.append((bag, torch.tensor(label)))
```

## Step-7 : Attention learns which frames matter

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class attentionMIL(nn.Module):
    def __init__(self, feature_dim = 512, hidden_dim=128):
        super().__init__()
        self.attention = nn.Sequential(
            nn.Linear(feature_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim,1)
        )

        self.classifier = nn.Sequential(
            nn.Linear(feature_dim, 1),
            nn.Sigmoid()
        )

    def forward(self, bag): 
        scores = self.attention(bag) #(num_frames,1)
        weights = F.softmax(scores, dim=0)

        #combines all frames into one vector
        clip_feature = torch.sum(weights * bag, dim=0)

        y = self.classifier(clip_feature)         # (1)
        return y, weights
```

# Step-8 : Training Loop

```python
model = attentionMIL().to(device)
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
num_epochs = 25

for epoch in range(num_epochs):
    total_loss = 0

    for bag, label in final_mil_dataset:
        bag = bag.to(device)
        label = label.float().unsqueeze(0).to(device)

        optimizer.zero_grad()

        output, attention = model(bag)
        loss = criterion(output, label)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}] Loss: {total_loss:.4f}")
```

```python
#Saving weight
torch.save(model.state_dict(), "attention_mil.pth") #trained weights & biases
```

# Prediction

```python
def predict_clip(frame_paths):
    model.eval()

    bag = extract_clip_features(frame_paths).to(device)

    with torch.no_grad():
        prob, attention = model(bag)

    pred = 1 if prob.item() > 0.5 else 0
    return prob.item(), pred, attention
```

```python
def extract_clip_features_from_faces(face_images):
    """
    face_images: numpy array (T, H, W, 3)
    returns: torch tensor (T, 512)
    """
    features = []

    with torch.no_grad():
        for face in face_images:
            img = Image.fromarray(face).convert("RGB")
            img = transform(img).unsqueeze(0).to(device)

            feat = resnet(img)        # (1,512)
            features.append(feat.squeeze(0))

    return torch.stack(features)
```
