```python
##variable length arguments
##positional and keywords arguments

def print_numbers(*krish):
    for number in krish:
        print(number)
    
##calling the function
print_numbers(1,2,3,4,5,6,7,8,"krish") 
       
```

```python
##positional argsuments
def print_numbers(*args):
    for number in args:
        print(number)
        
###calling the function
print_numbers(1,2,3,4,5,6.12,5.43,"jyoti")        
        
        
        
```

```python
##keyword args -> all the keys are in the form of key,values pair
def print_details(**kwargs):
    for key,values in  kwargs.items():
        print(f"{key}:{values}")
        
##calling the function   

print_details(name="jyoti",age=20,branch="cse(ai)")    
    
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"positional arguments:{val}")
        
        
    for key,values in  kwargs.items():
        print(f"{key}:{values}")
        
        
##calling the function
print_details(1,2,3,4,"jyoti",name="jyoti",age=20,branch="cse(ai)")
      
        
        
        
```
