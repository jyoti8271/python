File operation-read and write files

file handling is a crucial part of any programming language.python provides built in functions and methods to read from and write to files ,both text and binary.this lesson will cover the basics of the file handling,including reading and writing text files and binary files

```python
##read a whole file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## read a file line by line

with open('example.txt','r')as file:
    for line in file:
        print(line.strip()) #.strip() remove the new line character
        
```

```python
###open a file to writing a file (overwriting) problem

with open('example.txt','w')as file:
    file.write('hello world,\n')
    file.write("this is new line")
```

```python
## write a file without overwritting
with open('example.txt','a') as file:
    file.write("append operation taking place \n")
```

```python
### writting a list of lines to a file
lines=['first line\n','second line \n','third line \n']
with open('example.txt','a')as file:
    file.writelines(lines)
```

```python
### binary filess

# writing to binary files
data= b'\x00\x01\02\x03\x04'
with open('example.bin','wb')as file:
    file.write(data)
    
```

```python
##reading a file
with open('example.bin','rb')as file:
    content=file.read()
    print(content)
    
```

```python
## read the content from source text file and write to a destination text file
## copying a text file

with open('example.bin','r') as source_file:
    content=source_file.read()
    
with open('destination.bin','w') as destination_file:
    destination_file.write(content)    
```
