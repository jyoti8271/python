Introduction to dictionaries


Dictonaries are unordered collection of items .they store data in keys-value pairs.keys must be unique and immutable(e.g,strings,numbers,or tuples),while values can be of any type.

```python
###creating dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"krish","age":32,"grade":34}
print(student)
print(type(student))
```

```python
## Error
#single key error is always used
student={"name":"krish","age":32,"name":34}
print(student)
```

```python
###acessing dictionary elements
student={"name":"krish","age":32,"grade":"A"}
print(student)
print(student['grade'])
print(student['age'])
```

```python
##acessing using get method
print(student.get("grade"))
print(student.get("last_name"))
print(student.get("last name","not availabel"))
```

```python
###Modifying dictionary elements
##Dictionary are mutable,so you can add ,update or delete elements
print(student)
```

```python
student["age"]=33##updated the value for the key
print(student)###added a new key and value
```

```python
##delete th e key
del student["age"]
print(student)
```

```python
#####Dictionary methods

keys=student.keys()###get all the keys
print(keys)

value=student.values()### get all the values
print(value)

items=student.items()  ####get all the key and value pair
print(items)
```

```python
##### shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
###iterating over dictionary
###you can use loops to iterative over dictionaries,keys,values,itmes
##iterating over keys
for keys in student.keys():
    print(keys)
```

```python
##iterating over values
for values in student.values():
    print(values)
```

```python
###iterate over key value pairs 
for keys,values in student.items():
   print(f"{keys}:{values}")
```

```python
###nested dictionaries is 
student={
    "student1":{"name":"krish","age":32},
    "student2":{"name":"spiders","age":42}
}
print(student)



```

```python
##acesss nested dictionary elements
print(student["student2"]["name"])
print(student["student2"]["age"])
```

```python
student.items()
```

```python
###iterating over nested dictionaries
for student_id, student_info in student.items():
    print(student_id,student_info)
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
###dictionary comprehension
square={x:x**2 for x in range(5)}
print(square)
```

```python
###conditional dictional comprehension
even={x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
###practical examples

##use a dictionary to count the frequency of elements in list
number=[2,3,4,3,4,5,4,3,4,4,4]
frequency={}
for num in number:
    if num in frequency:
        frequency[num]+=1
    else:
        frequency[num]=1    
print(frequency)        
    
```

```python
##merge 2 dictionary into one
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
