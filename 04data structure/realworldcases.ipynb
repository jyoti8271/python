List are one of the most commonly used data structures in python ,thanks to their versality and ease of use.
here are several practical examples that illustrate their use in real-world scenarios

Example 1....Create a to do list to keep track of tasks

```python
to_do_list=["buy groceries","clean the house","pay bills"]

#adding the task
to_do_list.append("schedule meeting")
to_do_list.append("go for a run")

##removing a completed task
to_do_list.remove("clean the house")

##checking if the task in list
if "pay bills" in to_do_list:
    print("don't foreget to pay the utility bills")
print("to do list remaining task")

for work in to_do_list:
    print(work)
       
```
