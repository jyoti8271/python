sets

sets are a built in a data type in python used to store collections of unique items.they are unordered ,meaning that the elemets do no follow a specific order,and they do no allow duplicate elemnts.sets are useul for mombership tests,eliminating duplicate entries,and performing mathematical set operations like union,intersections,differrence and symmetric diffrence.

```python
##create a set
my_set={1,2,3,4,50}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_empty_set=set([1,2,3,4,5,6])
print(type(my_empty_set))
print(my_empty_set)
```

```python
##remove duplicate value
my_empty_set=set([1,2,3,4,5,6,5,6])
print(my_empty_set)
```

```python
##Basic set operation
##adding and removing elements
my_set.add(7)
print(my_set)

my_set.add(7)
print(my_set)
```

```python
##remove the elements from the set
my_set.remove(7)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(10)
print(my_set)
```

```python
#pop method usinf fifo first output will be always clear
my_set.pop()
print(my_set)
```

```python
removed_elements=my_set.pop()
print(removed_elements)
print(my_set)
```

```python
##clear all the elements
my_set.clear()
print(my_set)
```

```python
##set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
##mathematical opertations
set1={1,2,3,4,6}
set2={5,6,7,8,6}

##union
union_set=set1.union(set2)
print(union_set)



#intersection
intersection_set=set1.intersection(set2)
print(intersection_set)


set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,6}
set2={5,6,7,8,6}

#difference
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
##symmetric difference
set1.symmetric_difference(set2)
```

```python
##set methods
set1={1,2,3,5}
set2={3,4,5}

#  is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lt=[1,2,2,3,4,4,5]
set(lt)
```
